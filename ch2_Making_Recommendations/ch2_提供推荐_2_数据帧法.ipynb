```python
#定义两个目标之间的相似系数函数
from math import sqrt
def my_sim_pearson(df_pred,subj_column,subj1,subj2,obj_column):
    #df_pred: 偏好评价数据集
    #subj：比较主体，计算相似度的目标类别，人or物,如person/item
    #target1,target2：两个目标
    #obj：比较客体，如item/person
    
    df1 = df_pred[df_pred[subj_column] == subj1]
    df2 = df_pred[df_pred[subj_column] == subj2]
    
    #取两个目标之间的交集
    df_same = pd.merge(df1,df2,on=[obj_column])
    n = df_same.shape[0]
    
    #如果两个比较的主体之间没有共同点，则返回0
    if n == 0: return 0
    
    sum1 = df_same['mark_x'].sum()
    sum2 = df_same['mark_y'].sum()

    sum1Sq = pow(df_same['mark_x'],2).sum()
    sum2Sq = pow(df_same['mark_y'],2).sum()

    pSum = (df_same['mark_x'] * df_same['mark_y']).sum()

    num = pSum - (sum1 * sum2/n)
    den = sqrt((sum1Sq - pow(sum1,2)/n) * (sum2Sq - pow(sum2,2)/n))

    r = num/den
    
    return r
```

```python
def my_sim_distance(df_pred,subj_column,subj1,subj2,obj_column):
    df1 = df_pred[df_pred[subj_column] == subj1]
    df2 = df_pred[df_pred[subj_column] == subj2]
    
    #取两个目标之间的交集
    df_same = pd.merge(df1,df2,on=[obj_column])
    n = df_same.shape[0]
    
    #如果两个比较的主体之间没有共同点，则返回0
    if n == 0: return 0
    
    
    sum_of_squares = pow(df_same['mark_x'] - df_same['mark_y'],2).sum()
    return 1/(1+sqrt(sum_of_squares))
```

```python
#基于上一个函数计算所有比较主体之间的相似系数
def my_sim_all(df_pred,subj_column,obj_column,similar = my_sim_pearson):
    
    
    #初始化所有比较主体相似度数据容器dataframe
    s1 = subj_column+'_1'
    s2 = subj_column+'_2'
    df_sim_all = pd.DataFrame(columns=[s1,s2,'similarity'])
    #各个比较主体列表
    subjects = df_pred[subj_column].unique()
    
    #所有主体两两计算相似度系数
    for i in range(len(subjects)):
        if i < len(subjects) - 1 :
            for j in range(i+1,len(subjects)):
                score = similar(df_pred, subj_column,subjects[i],subjects[j],obj_column)#计算一组本体之间的相似度系数
                df_sim_all = df_sim_all.append([{s1:subjects[i],s2:subjects[j],'similarity':score}],ignore_index=True)#添加到dataframe中
    
    #对填充后的dataframe进行降序排序
    df_sim_all = df_sim_all.sort_values(by = [s1,'similarity'], ascending=False)
    return df_sim_all
```

```python
#制定函数
def getRecommendations(df_pred,df_sim_all,subj,subj_entity,obj):
    #df_pred, 原始偏好库
    #df_sim_all, 所有主体之间关联相似度
    #subj,主体名称
    #subj_entity,特定主体
    #obj，比较客体名称
    
    
    #筛选相关系数>0
    subj1 = subj + '_1'
    subj2 = subj + '_2'
    df_subj_entity = df_sim_all[((df_sim_all[subj1] == subj_entity) | (df_sim_all[subj2] == subj_entity)) & (df_sim_all['similarity'] > 0)]
    
    
    
    #找到subj没有而other有的obj
    list_obj_subj_entity = list(df_pred[df_pred[subj] == subj_entity][obj].unique())
    list_obj_all = list(df_pred[obj].unique())
    unseen_obj = list(set(list_obj_all).difference(set(list_obj_subj_entity)))          
    
            
    for un in unseen_obj:#遍历没看过的
        df_subj_entity[un] = np.NaN
        w_un = 'weight_'+un
        df_toby[w_un] = np.NaN 
        other = None
                                
        for ix,r in df_subj_entity.iterrows():#遍历other
            #确定与谁比较
            if r[subj1] != subj_entity:
                other = r[subj1]
            else:
                other = r[subj2]
                
        #找到other看过但toby没看过的评分
            if un in list(df_pred[df_pred[subj] == other][obj]):   
                df_subj_entity.at[ix,un] = df_pred[(df_pred[subj] == other) & (df_pred[obj] == un)]['mark'].values[0]
            
    #计算权重参数
        df_subj_entity[w_un] = df_subj_entity['similarity'] * df_subj_entity[un]
                                
    
    #初始化推荐矩阵
    df_recommendations_entity = pd.DataFrame(columns=[subj,obj,'recom_mark'])

    for un in unseen_obj:
        w_un = 'weight_'+un
        recom_mark = df_subj_entity[w_un].sum()/df_subj_entity[df_subj_entity[w_un].notnull()]['similarity'].sum()
        df_recommendations_entity = df_recommendations_entity.append([{'person':subj_entity,'item':un,'recom_mark':recom_mark}],ignore_index=True)
    
    df_recommendations_entity = df_recommendations_entity.sort_values(by = ['recom_mark'], ascending=False)#排序
                         
    
    return df_recommendations_entity
```

```python
#得到所有主体的推荐参数
def getRecommendations_All(df_pred,df_sim_all,subj,obj):
    
    #为所有主体的推荐指数设置一个初始化datarframe
    df_recommendations_all = pd.DataFrame(columns=[subj,obj,'recom_mark'])
    
    for s in df_pred[subj].unique():
        df_recommendations_all = df_recommendations_all.append(getRecommendations( df_pred, df_sim_all, subj, s, obj),ignore_index=True)
    
    return df_recommendations_all
```

```python
def getRecommendations_from_items_sim(df_pred,df_sim_all_item,person_entity):
   
    df_person_entity = df_pred[df_pred['person'] == person_entity]
    
    seen_list = list(df_critics[df_critics['person'] == person_entity]['item'])
    all_list = list(df_critics['item'].unique())
    #找到特定人没看过的片子
    unseen = list(set(all_list).difference(set(seen_list)))
    
    #遍历没看过的
    for un in unseen:
                  
        df_toby[un] = np.NaN
        w_un = 'weight_'+un
        df_toby[w_un] = np.NaN 
        other = None
    
        for ix,r in df_toby.iterrows():#遍历每一行
        
            seen = df_toby.at[ix,'item']
            #获取unseen和seen两部片子之间的相似度
            if seen == un:continue
            sim = df_sim_all_item[(df_sim_all_item['item_1'] == seen) & (df_sim_all_item['item_2'] == un) | (df_sim_all_item['item_1'] == un) & (df_sim_all_item['item_2'] == seen)]['similarity'].values[0]
            #如果相似度<0，则丢弃
            if sim < 0:continue
            df_toby.at[ix,un] = sim
            
        #计算权重参数
        df_toby[w_un] = df_toby['mark'] * df_toby[un]
    
    #初始化推荐矩阵
    df_recommendations = pd.DataFrame(columns=['person','item','recom_mark'])

    for un in unseen:
        w_un = 'weight_'+un
        recom_mark = df_toby[w_un].sum()/df_toby[un].sum()
        df_recommendations = df_recommendations.append([{'person':person_entity,'item':un,'recom_mark':recom_mark}],ignore_index=True)
        
    df_recommendations = df_recommendations.sort_values(by = ['recom_mark'], ascending=False)#排序

    
    return df_recommendations.dropna()
```

```python
#得到所有主体的推荐参数
def getRecommendations_from_items_sim_All(df_pred,df_sim_item):
    
    #为所有主体的推荐指数设置一个初始化datarframe
    df_recommendations_all = pd.DataFrame(columns=['person','item','recom_mark'])
    
    for p in df_pred['person'].unique():
        
        df_recommendations_all = df_recommendations_all.append(getRecommendations_from_items_sim(df_pred,df_sim_item,p),ignore_index=True)
    
    return df_recommendations_all
```
