```python
#对教材中的princlust做如下修改
def printclust(clust, labels=None,n=0):
    for i in range(n): print(' ',end='')#注意打印空格后不要换行
    if clust.id<0:
        print('-')#表示一级枝节点
    else:
        if labels==None: print(clust.id)
        else: print(labels[clust.id])#表示叶节点，即某个blogname
    
    if clust.left!=None: printclust(clust.left,labels=labels,n=n+1)
    if clust.right!=None: printclust(clust.right,labels=labels,n=n+1)
```

```python
#修改原代码中给的第一行 在print函数中增加end=''，在python中，默认print()执行后会换行，因此需要end来取消换行
printclust(clust,labels = blognames)
```

#### 假设将该样本集分为5类

```python
clusters.kcluster(data,k=5)
```

```python
#将以ID为展现的聚类结果转换为以title为展现的聚类结果
def displayClusterTitle(bestmatches,rownames):
    #参数1：聚类结果列表
    #参数2：样本ID对应名称
    
    title = [ [] for kid in range(len(bestmatches)) ]
    
    #遍历每一个聚类
    for kid in range(len(bestmatches)):
        #遍历每一个聚类中的样本：
        for sampleid in bestmatches[kid]:
            #添加样本标题
            title[kid].append(rownames[sampleid])
            
    return title
```

```python
displayClusterTitle(clusters.kcluster(data,k=5), blognames)
```

#### 定义缩放函数

```python
#输入参数1 data，样本数据集
#输入参数2 distance，相似度距离计算函数
#输入参数3 rate，调整因子

def scaledown(data, distance = clusters.pearson, rate = 0.01):
    #样本数
    n = len(data)
    
    #计算两两样本之间真实距离，样本和样本之间的距离可以构成一个 n x n的矩阵，这里的距离采用的是pearson距离计算出来的*
    realdist = [[distance(data[i],data[j]) for j in range(n)] for i in range(n)]
    
    outersum = 0.0
    
    #初始距离，随机初始化两两样本之间的距离
    #先随机初始化n个样本的位置
    loc = [[random.random(),random.random()] for i in range(n)]
    #再初始化点位之间的距离，为0
    fakedist = [[0.0 for j in range(n)] for i in range(n)]
    
    lasterror = None
    
    #默认最多调整1000次，即预计1000次以内可以找到与目标距离最接近的位置
    for m in range(0,1000):
        #遍历距离矩阵
        for i in range(n):
            for j in range(n):
                #先计算随机初始化两两样本之间的距离,采用空间几何距离法计算*
                fakedist[i][j] = sqrt(sum([pow(loc[i][x] - loc[j][x],2)]) for x in range(len(loc[i])))
                
        #移动节点
        #x轴/y轴初始化移动距离,也是一个nxn的矩阵
        grad = [[0.0,0.0] for i in range[n]]
        
        totalerror  = 0
        
        #计算误差距离矩阵
        #遍历两两样本
        for k in range(n):
            for j in range(n):
                if j == k: continue
                    #计算出真实距离与当前距离之间的误差之比
                errorterm = (fakedist[j][k] - realdist[j][k])/realdist[j][k]
                    
                    #根据相对距离误差，等比例计算出k点和j点之间x轴/y轴方向需要调整的距离
                grad[k][0] += ((loc[k][0] - loc[j][0])/fakedist[j][k]) * errorterm
                grad[k][1] += ((loc[k][1] - loc[j][1])/fakedist[j][k]) * errorterm
                    
                totalerror += abs(errorterm)
        print(totalerror)
    
        #如果当前总误差比上一次总误差还要大，则退出，不再计算
        if lasterror and lasterror < totalerror:break
        
        lasterror = totalerror
        
        #根据计算出来的移动距离，更新坐标
        for k in range(n):
            loc[k][0] -= rate*grad[k][0]
            loc[k][1] -= rate*grad[k][1]
                                   
    return loc
```
