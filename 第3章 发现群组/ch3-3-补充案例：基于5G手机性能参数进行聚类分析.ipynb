```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.rcParams['font.sans-serif'] = ['SimHei']

price = list(df_all['price'])
camera = list(df_all['camera'])

plt.scatter(price,camera)
note = list(df_all['cellphone'])

for i in range(len(note)):
    plt.annotate(list(note)[i], xy = (price[i], camera[i]), xytext = (price[i]+0.5, camera[i]+0.5),fontsize = 15) # 这里xy是需要标记的坐标，xytext是对应的标签坐标


plt.xlabel('price',fontsize = 15)
plt.ylabel('camera',fontsize = 15)
plt.title('散点图：手机价格与相机评分',fontsize = 20)

#分割象限
# plt.plot([0.5,0.5],[0,0.7], linewidth=4, color='red',linestyle="--" )
# plt.plot([0,1],[0.33,0.33], linewidth=4, color='red',linestyle="--")

plt.show()
```

## 3、使用k-均值聚类分析，将样本集分为四部分

```python
data = df_all[['price','camera']].values
```

```python
print(data)
```

```python
result = clusters.kcluster(data,k = 4)
for r in result:
    if len(r) > 0 :
        cellphone = []
        for i in r:
            cellphone.append(df_all['cellphone'][i])
        print(cellphone)
```
